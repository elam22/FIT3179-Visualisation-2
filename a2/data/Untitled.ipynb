```python
import pandas as pd
```

```python
df = pd.read_csv("whc-sites-2019.csv")
df.head()
```

```python
set(df['states_name_en'])
```

```python
#count number of sites by country
country_lst = {}
for index, row in df.iterrows():
    state = row['states_name_en'].split(',')
    if len(state)>1:
        for item in state:
            if item in country_lst:
                country_lst[item] = country_lst[item] +1
            else:
                country_lst[item] = 1
    else:
        state = state[0]
        if state in country_lst:
            country_lst[state] = country_lst[state] +1
        else:
            country_lst[state] = 1
```

```python
(pd.DataFrame.from_dict(data=country_lst, orient='index')
   .to_csv('country_count.csv', header=['count']))

cc = pd.read_csv('country_count.csv').rename(columns={"Unnamed: 0": "Country"})
cc.head()
```

```python
cd = pd.read_csv('country_code.csv')
cd.head()
```

```python
for index, row in cd.iterrows():
    index_cc = list(cc[cc['Country']==row['name']].index.values)
    if index_cc:
        index_cc = index_cc[0]
        cd.loc[index, 'country_count'] = str(cc.loc[index_cc, 'count'])
cd.head()
```

```python
cd.to_csv('country_count_total.csv', index=False)
```

```python
df['date_inscribed']
```

```python
top_countries = cc.sort_values(by=["count"], ascending=False)[0:10]
top_countries
```

```python
for index, row in df.iterrows():
    if row['states_name_en'] in list(top_countries['Country']):
        df.loc[index, 'yes'] = True
```

```python
dout = df[df['yes']==True]
```

```python
for index, row in df.iterrows():
    state = row['states_name_en'].split(',')[0]
    ind = list(cd[cd['name'] == state].index.values)
    if ind:
        ind = ind[0]
        region = cd.loc[ind, 'region']
        if region == 'Africa':
            df.loc[index, 'region'] = 'Africa'
        elif region == 'Asia':
            df.loc[index, 'region'] = 'Asia'
        elif region == 'Oceania':
            if 'Australia' in row['states_name_en']:
                df.loc[index, 'region'] = 'Australia'
            else:
                df.loc[index, 'region'] = 'Oceania'
        elif region == 'Europe':
            df.loc[index, 'region'] = 'Europe'
        elif region == 'Americas':
            subregion = cd.loc[ind, 'sub-region']
            if subregion == 'Latin America and the Caribbean':
                df.loc[index, 'region'] = 'South America'
            elif subregion == 'Northern America':
                df.loc[index, 'region'] = 'North America'
            else:
                df.loc[index, 'region'] = 'nan'
        else:
             df.loc[index, 'region'] = 'nan'
    else:
        df.loc[index, 'region'] = 'nan'
```

```python
region_lst = ['Asia','Europe','Africa','South America','Australia','North America','Oceania']
region_year_lst = []
for region in region_lst:
    year_lst = {}
    for index, row in df.iterrows():
        if row['region'] == region:
            year = row['date_inscribed']
            if year in year_lst:
                 year_lst[year] = year_lst[year] +1
            else:
                year_lst[year] = 1
    region_year_lst.append(year_lst)
```

```python
for i in range(len(region_lst)):
    region_lst_min = region_year_lst[i]
    (pd.DataFrame.from_dict(data=region_lst_min, orient='index').to_csv(str(region_lst[i]) + '_count.csv', header=['count']))
    yc = pd.read_csv(str(region_lst[i]) + '_count.csv').rename(columns={"Unnamed: 0": "Year"})
    yc = yc.sort_values(by=["Year"])
    yc['count_cumsum'] = yc['count'].cumsum()
    yc['Region'] = str(region_lst[i])
    yc.to_csv(str(region_lst[i]) + '_year_count.csv', index=False)
```
