```python
import pandas as pd
```

```python
df = pd.read_csv("whc-sites-2019.csv")
df.head()
```

```python
set(df['category'])
```

```python
set(df['states_name_en'])
```

```python
set(df['region_en'])
```

```python
#count number of sites by country & region
country_lst = {}
region_lst = {}
for index, row in df.iterrows():
    state = row['states_name_en'].split(',')
    if len(state)>1:
        for item in state:
            if item in country_lst:
                country_lst[item] = country_lst[item] +1
            else:
                country_lst[item] = 1
    else:
        state = state[0]
        if state in country_lst:
            country_lst[state] = country_lst[state] +1
        else:
            country_lst[state] = 1
    region = row['region_en'].split(',')
    if len(region)>1:
        for item in region:
            if item in region_lst:
                region_lst[item] = region_lst[item] +1
            else:
                region_lst[item] = 1
    else:
        region = region[0]
        if region in region_lst:
            region_lst[region] = region_lst[region] +1
        else:
            region_lst[region] = 1
```

```python
(pd.DataFrame.from_dict(data=country_lst, orient='index')
   .to_csv('country_count.csv', header=['count']))
```

```python
(pd.DataFrame.from_dict(data=region_lst, orient='index')
   .to_csv('region_count.csv', header=['count']))
```
